0 <= x1,x2 <= pi


Importing libraries 
>inspyred is a free, open source framework for creating biologically-inspired computational intelligence algorithms in Python, including evolutionary computation, swarm intelligence, and immunocomputing. Additionally, inspyred provides easy-to-use canonical versions of many bio-inspired algorithms for users who do not need much customization

```python
from __future__ import division
import inspyred
import random
import matplotlib.pyplot as plt
import math
from inspyred import ec
from inspyred.ec import terminators
```

```python
# Return a random floating point number N such that a <= N <= b for 0 <= 3.14.
#This function must take the random number generator object along with the keyword arguments. 
#Notice that we can use the args variable to pass anything we like to our functions

def pseudo_generator(random, args):
    return [random.uniform(0.0000000000000001, 3.14) for _ in range(2)]
```

```python
#Creating constraints for (x1)^2 + (x2)^2 -1 0.1 cos(16arctan(x1/x2)) >= 0 ; (x1-0.5)^2 + (x2-0.5)^2 <= 0.5
#Return the fitness value
```

```python
@inspyred.ec.evaluators.evaluator
def constraints(objective, args):
    x1 = objective[0]
    x2 = objective[1]
    constraints1 = x1 ** 2 + x2 ** 2 - 1 - 0.1 * math.cos(16*math.atan(x1/x2))
    constraints2 = (x1 - 0.5) ** 2 + (x2 - 0.5) ** 2
    if constraints1 < 0:
        x1 += 3.14 + abs(constraints1)
        x2 += 3.14 + abs(constraints1)
    if constraints2 > 0.5:
        x1 += 3.14 + (constraints2 - 0.5)
        x2 += 3.14 + (constraints2 - 0.5)
    return inspyred.ec.emo.Pareto([x1, x2])
```

```python
#Return the next random floating point number in the range randseed initiated.
#candidate solutions are [0,3,14) range strings whose decimal values should be minimized: under constraints 
#A variator that determines how offspring are created from existing individuals
#A terminator that determines whether the evolution should end
#An observer that defines how the user can monitor the state of the evolution
#Notice that this is essentially a Gaussian mutation on each coordinate of each tuple. Now we can create our custom EC.
#This EC uses tournament selection, uniform crossover, our custom mutation operator, and steady-state replacement.


def evolutionary_computation():
    randseed = 12345
    prng = random.Random(randseed)
    ea = inspyred.ec.emo.NSGA2(prng)
    ea.variator = [inspyred.ec.variators.blend_crossover,
                   inspyred.ec.variators.gaussian_mutation]
    ea.terminator = inspyred.ec.terminators.evaluation_termination
    ea.observer = [inspyred.ec.observers.archive_observer]
    final_pop = ea.evolve(generator=pseudo_generator,
                          evaluator=constraints,
                          pop_size=100,
                          num_selected=100,
                          tournament_size=10,
                          maximize=False,
                          bounder=inspyred.ec.Bounder(0.0000000000000001, 3.14),
                          max_evaluations=5000)
    display = True
    if display:
        final_arc = ea.archive
        print('Optimal Solutions: \n')
        for f in final_arc:
            print(f)
        x = []
        y = []
        for f in final_arc:
            x.append(f.fitness[0])
            y.append(f.fitness[1])
        plt.scatter(x, y, color='r')
        plt.xlabel("F(x) 1")
        plt.ylabel("F(x) 2")
        
        plt.show()

evolutionary_computation()
```

This module implements the Sequential Least SQuares Programming optimization
algorithm (SLSQP)

```python
import numpy as np
from scipy.optimize import minimize
np.random.seed(1)
#from math import cos
#from math import atan
#from math import pi
import math

xw = []
yw = []
for j in np.arange(0, 1, 0.0005):
    fun = lambda x: (x[0])*j + (x[1])*(1-j)
    cons = ({'type': 'ineq', 'fun': lambda x:  -(x[0])**2 - (x[1])**2 + 1 + 0.1*math.cos(16*(math.atan(x[0]/x[1])))},
            {'type': 'ineq', 'fun': lambda x:  (x[0] - 0.5)**2 + (x[1] - 0.5)**2 - 0.5})

    bnds = ((0.00001, 3.14), (0.00001, 3.14))

    res = minimize(fun, (0.75, 0.75), method='SLSQP', bounds=bnds,
               constraints=cons, options = {'maxiter': 10000})
    xw.append(res.x[0])
    yw.append(res.x[1])
    #print(j, 1-j, res.x)
    
plt.scatter(xw, yw, color='r')
plt.xlabel("F(x1)")
plt.ylabel("F(x2)")
plt.show()
```
